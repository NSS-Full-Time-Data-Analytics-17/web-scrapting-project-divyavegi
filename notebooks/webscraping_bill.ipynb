```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
```

#### 1. Navigate to https://www.billboard.com/charts/hot-100/. Using BeautifulSoup, extract out the This Week, artist, song, Last Week, Peak Position, and Weeks on Chart values into a pandas DataFrame. Hint: The HTML for the number one ranked song is slightly different from that of the rest of the songs.


```python
url = 'https://www.billboard.com/charts/hot-100/'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}

response = requests.get(url, headers=headers)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

```python
soup.find('li')
```

```python
songs = soup.findAll("div",class_="o-chart-results-list-row-container")
songs
```

```python
data = []
for s in songs:
    title = s.find("h3")
    artist = s.find("span", class_="c-label")
    if not title or not artist:
        continue
    title = title.get_text(strip=True)
    artist = artist.get_text(strip=True)
    last = s.find("span", {"data-label": "Last Week"})
    peak = s.find("span", {"data-label": "Peak Position"})
    weeks = s.find("span", {"data-label": "Weeks on Chart"})
    data.append({"Song": title,"Artist": artist,
        "Last Week": last.get_text(strip=True)if peak else None,
        "Peak Position": peak.get_text(strip=True) if peak else None,
        "Weeks on Chart": weeks.get_text(strip=True) if weeks else None })
song_df = pd.DataFrame(data)
song_df
```
